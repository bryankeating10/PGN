```python
import pandas as pd

# PGN file path
pgn_path = "Morphy.pgn"

game_meta = []
moves = []

with open(pgn_path) as pgn:
    game_id = 1 # Index the games by game ID

    while True:
        game = ch.read_game(pgn)

        if game is None:
            break # End of PGN file

        # Extract metadata
        headers = dict(game.headers)
        headers["Game_ID"] = game_id
        games_meta.append(headers)
        game_id += 1

game_df = pd.DataFrame(games_meta)
game_df.set_index('Game_ID',inplace=True)
```
